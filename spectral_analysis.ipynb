```python
from numpy import genfromtxt
import matplotlib.pyplot as plt
import numpy as np
```

```python
samples = genfromtxt('test.csv', delimiter=',')
```

```python
clock = 44100
frequency = 440
t = np.arange(samples.size)/clock

start, stop = 0, 1000
plt.figure(figsize = (15, 3), dpi = 100)
plt.plot(t[start:stop], samples[start:stop], linewidth = 1, c = "k")
```

```python
timer_period = int(np.floor(clock / (frequency * 8.0)))
real_frequency = clock/(8*timer_period)
```

```python
plt.figure(figsize = (10, 3), dpi = 100)
fft = np.fft.fft(samples)
freq = np.fft.fftfreq(samples.size, 1/clock)
spectrum = 10*np.log10(np.abs(fft))
plt.plot(freq, spectrum, c = "k", linewidth = 1)
plt.xlim([0, 2500])
plt.ylim([0, spectrum.max()+ 5])
plt.vlines(x=np.arange(10)*frequency, ymin = 0, ymax = spectrum.max()+5, color = "blue", linewidths = 1, label = "target frequency")
plt.vlines(x=np.arange(10)*real_frequency, ymin = 0, ymax = spectrum.max()+5, color = "red", linewidths = 1, label = "true frequency")
plt.grid(linewidth = 0.5, c = "k", linestyle = "--", alpha = 0.4)
plt.legend()
```
